# Gaussian Distribution

For any gaussian distribution, we need to specify the mean value ($\mu$) and the standard deviation ($\sigma$).

If we measure a physical quantity with the values of $\mu$ and $\sigma$
as specified, we can get ANY value between $+\infty$ and $-\infty$.

In practice, almost all of the probability distribution is contained
in the region between $\mu-6\sigma$ and $\mu+6\sigma$.



```python
# import the required packages with standard naming convention
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

mu = 100
sigma = 10

x = np.arange(mu-6.0*sigma,mu+6.0*sigma,0.01)
```

The probability density function (pdf) gives the actual probability
of measuring the value, x.

```python
gaussian_pdf = stats.norm.pdf(x,mu,sigma)

plt.plot(x,gaussian_pdf)
plt.title('Gaussian PMF: mu=%i, sigma=%.2f' % (mu,sigma), fontsize=15)
plt.xlabel('X')
plt.ylabel('Probability',fontsize=15)
plt.show()
```

The cumulative distribution function (cdf) gives the 
INTEGRATED probability from the lowest allowed x-value (in this case
x = -$\infty$) as a function of x.

```python
gaussian_cdf = stats.norm.cdf(x,mu,sigma)

plt.plot(x,gaussian_cdf)
plt.title('Gaussian CDF: mu=%i, sigma=%.2f' % (mu,sigma), fontsize=15)
plt.xlabel('X')
plt.ylabel('Integrated Probability',fontsize=15)
plt.show()
```

The percentile point function (ppf) is the INVERSE of the cdf. If we specify the integrated probability (y-axis of the cdf plot), it will tell us what x-axis value that corresponds to.

```python
alpha = np.arange(0.0,1.0,0.01)

gaussian_ppf = stats.norm.ppf(alpha,mu,sigma)

plt.plot(alpha,gaussian_ppf)
plt.title('Gaussian PPF: mu=%i, sigma=%.2f' % (mu,sigma), fontsize=15)
plt.xlabel('Integrated Probability')
plt.ylabel('X',fontsize=15)
plt.show()
```
