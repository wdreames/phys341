# Student's t distribution

For any t distribution, we specify only the number of degrees of freedom ($\nu$).  Note that for $\nu \rightarrow \infty$, the t-distribution approaches a Gaussian distribution with $\mu = 0$ and $\sigma = 1$.

In practice, for the t-distibution, almost all of the probability 
distribution is contained in the region between -6 and 6.



```python
# import the required packages with standard naming convention
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

ndof1 = 3
ndof2 = 1000

x = np.arange(-6.0,6.0,0.01)
```

The probability density function (pdf) gives the actual probability of measuring the value, x.

```python
t_pdf1 = stats.t.pdf(x,ndof1)
t_pdf2 = stats.t.pdf(x,ndof2)

plt.plot(x,t_pdf1,label='Ndof = %.0f' % ndof1)
plt.plot(x,t_pdf2,label='Ndof = %.0f' % ndof2)
plt.title('t PMF', fontsize=15)
plt.xlabel('X')
plt.ylabel('Probability',fontsize=15)
plt.legend(loc='best', frameon=False)
plt.show()
```

The cumulative distribution function (cdf) gives the INTEGRATED probability from the lowest allowed x-value (in this case x = -$\infty$) as a function of x.


```python
t_cdf1 = stats.t.cdf(x,ndof1)
t_cdf2 = stats.t.cdf(x,ndof2)

plt.plot(x,t_cdf1,label='Ndof = %.0f' % ndof1)
plt.plot(x,t_cdf2,label='Ndof = %.0f' % ndof2)
plt.title('t CDF', fontsize=15)
plt.xlabel('X')
plt.ylabel('Integrated Probability',fontsize=15)
plt.legend(loc='best', frameon=False)
plt.show()
```

The percentile point function (ppf) is the INVERSE of the cdf. If we specify the integrated probability (y-axis of the cdf plot), it will tell us what x-axis value that corresponds to.

```python
alpha = np.arange(0.0,1.0,0.01)

t_ppf1 = stats.t.ppf(alpha,ndof1)
t_ppf2 = stats.t.ppf(alpha,ndof2)

plt.plot(alpha,t_ppf1,label='Ndof = %.0f' % ndof1)
plt.plot(alpha,t_ppf2,label='Ndof = %.0f' % ndof2)
plt.title('t PPF', fontsize=15)
plt.xlabel('Integrated Probability')
plt.ylabel('X',fontsize=15)
plt.legend(loc='best', frameon=False)
plt.show()
```
