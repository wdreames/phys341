# Binomial Distribution

For any binomial distribution, we need to specify the total number of times the experiment is performed (N), and the probability of the desired/interesting outcome (p).

If an experiment is performed N times, the total number of times that the desired/interesting outcome happens (k) can be any value between 0 and N, inclusive.



```python
# import the required packages with standard naming convention
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

N = 12
p = 1/2.0

k = np.arange(0,N+1)
```

The probability mass function (pmf) gives the actual probability of the desired/interesting outcome, as a function of k.

```python
binomial_pmf = stats.binom.pmf(k,N,p)

plt.bar(k,binomial_pmf)
plt.title('Binomial PMF: n=%i, p=%.2f' % (N,p), fontsize=15)
plt.xlabel('Number of desired outcomes')
plt.ylabel('Probability',fontsize=15)
plt.show()
```

The cumulative distribution function (cdf) gives the INTEGRATED probability from the lowest allowed x-value (in this case k=0) as a function of x.

```python
binomial_cdf = stats.binom.cdf(k,N,p)

plt.bar(k,binomial_cdf)
plt.title('Binomial CDF: n=%i, p=%.2f' % (N,p), fontsize=15)
plt.xlabel('Number of desired outcomes')
plt.ylabel('Integrated Probability',fontsize=15)
plt.show()
```

The percentile point function (ppf) is the INVERSE of the cdf.  If we specify the integrated probability (y-axis of the cdf plot), it will tell us what x-axis value that corresponds to.

```python
alpha = np.arange(0.0,1.0,0.01)

binomial_ppf = stats.binom.ppf(alpha,N,p)

plt.plot(alpha,binomial_ppf)
plt.title('Binomial PPF: n=%i, p=%.2f' % (N,p), fontsize=15)
plt.xlabel('Integrated Probability')
plt.ylabel('Number of desired outcomes',fontsize=15)
plt.show()
```
